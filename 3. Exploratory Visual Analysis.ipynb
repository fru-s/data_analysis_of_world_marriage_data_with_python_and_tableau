### 3.  Exploring Relationships

```python
# creating correlation matrix
em_cm_div.corr()
```

#### Creating a correlation heatmap using seaborn

```python
# Create a subset without the years

sub_em_cm_div = em_cm_div[['Ever_Married_%','Currently_Married_%','Divorced_%']]
```

```python
sub_em_cm_div
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub_em_cm_div.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

#### There is rather a weak, but positive relationship between the % of divorced people and the other two values, meaning: even if the number of currently or even married people increases for a certain country and certain age group and certain period and certain sex, the number of divorced people will only a slighly increase for the same country, age group and period, but almost not at all. There is a strong positive relationship between the proportion of ever married and currently married people though (again for same sex, age group, period of time and country), which was expected, but still not a linear relationship.

#### Creating a scatterplot for the variables with the strongest correlations: ever married and currently married

```python
sns.lmplot(x = 'Currently_Married_%', y = 'Ever_Married_%', data = em_cm_div)
```

#### Currently married men and women are persons who have been married and are not either divorced, widowed or separated.
#### Ever married men and women are those who have been married at least once in their lives, regardless of their current marital status.

#### The positive correlation is indeed strong, a huge part of the data values follow the trend line, but there is another bigger group of points where the Ever_Married Values is between 80% and 100% and the currently married is between 10 and 100%. I am assuming these should be the older age groups when most people where married at least once, but currently they could be divorced, widowed, married etc.
