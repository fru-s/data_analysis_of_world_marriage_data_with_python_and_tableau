### 4. Data consistency Checks

```python
#Need to make sure that DataValue column isn't mixed
for col in ms.columns.tolist():
  weird = (ms[[col]].applymap(type) != ms[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (ms[weird]) > 0:
    print (col)
```

```python
#Checking on missing values
ms.isnull().sum()
```

#### Non-standard_AgeGroups aren't neccesary for the analysis, they are rather confusing because it is only 1-2% of the total dataset, and can't be compared to the rest of the dataset. Hence I will create a subset of the data, where Non-standard_AgeGroups are not included

```python
#Filtering only for standard age groups
ms_s = ms[ms['Non-standard_AgeGroups'].isnull() == True]
```

```python
ms.shape
```

```python
ms_s.shape
```

```python
#Filtering only for standard age groups
ms_s = ms[ms['Series_contains_Non-standard_AgeGroups'].isnull() == True]
```

```python
#now non standard age groups can be dropped
ms_s=ms_s.drop(columns=['Non-standard_AgeGroups','Series_contains_Non-standard_AgeGroups'])
```
